```python
import pandas as pd
```

```python
data_set= pd.read_csv('Phones.csv',index_col=0)
data_set.head()
```

```python
# ADD THIS NEW BLOCK
import numpy as np

# Find brands that appear less than 15 times in the dataset
brand_counts = data_set['Brand'].value_counts()
brands_to_group = brand_counts[brand_counts < 15].index

# Replace these rare brands with the label 'Other'
data_set.loc[data_set['Brand'].isin(brands_to_group), 'Brand'] = 'Other'

print("This is the new, cleaned distribution of brands:")
print(data_set['Brand'].value_counts())
```

```python

print(data_set.columns)
```

```python
data_set.drop('Touchscreen', axis=1, inplace=True)
```

```python
CTD=['Wi-Fi', 'Bluetooth', 'GPS','3G', '4G/ LTE']
data_set.drop(CTD, axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le =  LabelEncoder()
```

```python
to_be_encoded=['Name','Brand', 'Model','Operating system']
for col in to_be_encoded:
    data_set[col] = le.fit_transform(data_set[col])
```

```python
data_set.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
X = data_set[[ "Screen size (inches)", "Operating system", "RAM (MB)", "Internal storage (GB)", "Battery capacity (mAh)", "Front camera", "Rear camera", "Price", "Number of SIMs", "Resolution x", "Resolution y", "Processor"]]
```

```python
Y = data_set["Brand"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(y_train.shape)
print(X_train.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

# 1. Initialize the Decision Tree model
model = DecisionTreeClassifier(random_state=42)

# 2. Train the model on your scaled training data
model.fit(X_train, y_train)

# 3. Make predictions on the test data
# model
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_test, y_pred) 
print(cm)

accuracy = metrics.accuracy_score(y_test, y_pred) 
print("Accuracy score:",accuracy)


```
